# Strategy



1. Um exemplo disso, so as diferentes formas de dar um desconto para um cliente em uma loja
2. Cada forma de dar desconto ser diferente de acordo com cada um dos clientes
3. Assim, podem-se criar vrias estratgias para definir um desconto
4. Ou seja, as classes ficam abertas para a implementao de diversos mtodos diferentes de uma interface, mas elas no so modificadas

---

Esse mtodo pode ser usado tanto passando strategies na forma de funo quanto usando uma classe para as strategies

---


## Usando classes

1. Esse  o padro para a implementao do Strategy
2.  criada uma interface (classe abstrata, para o Python) e essa classe ir ser herdada por outras classes
3. Cada classe ser uma determinada strategy, que ir retornar algo especfico

```python
from abc import ABC, abstractmethod
from typing import List


class Representation:

    def __str__(self):
        params = [f'{k}={v}' for k, v in self.__dict__.items()]
        return f'{self.__class__.__name__}({f", ".join(params)})'

    def __repr__(self):
        return self.__str__()
```

```python
class Product(Representation):
    def __init__(self, name: str, price: float) -> None:
        self.name = name
        self._price = price

    @property
    def price(self) -> float:
        return self._price
```

Definio da classe abstrata que ir ser herdada por cada strategy especfica que estar no mtodo `DiscountStrategy.calculate()`

```python
class DiscountStrategy(ABC):

    @abstractmethod
    def calculate(self, value: float) -> float: pass
```

Definio de cada strategy

```python
class TwentyPercentOff(DiscountStrategy):

    def calculate(self, value: float) -> float:
        return value * 0.8


class FiftyPercentOff(DiscountStrategy):

    def calculate(self, value: float) -> float:
        return value * 0.5


class NoDiscount(DiscountStrategy):

    def calculate(self, value: float) -> float:
        return value


class CustumedDiscount(DiscountStrategy):

    def __init__(self, discount: float):
        self._discount = discount

    def calculate(self, value: float) -> float:
        return value * (1 - self._discount)
```

Definio da classe que ir usar as strategies, ou seja, a classe Contexto
A strategy ser utilizada ao se pedir a property `subtotal`

```python
class Order(Representation):

    def __init__(self, products: List[Product], discount: DiscountStrategy):
        self._total_value = sum([product.price for product in products])
        self._discount = discount

    @property
    def total(self):
        return self._total_value

    @property
    def subtotal(self):
        return self._discount.calculate(self._total_value)
```

```python
products = [
        Product('Nonebook Acer Linux', 500.0),
        Product('Keyboard Razer', 250.0),
        Product('Headphone Razer', 250.0),
    ]


order1 = Order(products, FiftyPercentOff())
order2 = Order(products, TwentyPercentOff())
order3 = Order(products, NoDiscount())
order4 = Order(products, CustumedDiscount(0.05))

print('Total 1', end=': ')
print(order1.total)
print('Total 2', end=': ')
print(order2.total)
print('Total 3', end=': ')
print(order3.total)
print('Total 4', end=': ')
print(order4.total)

print('Subtotal 1', end=': ')
print(order1.subtotal)
print('Subtotal 2', end=': ')
print(order2.subtotal)
print('Subtotal 3', end=': ')
print(order3.subtotal)
print('Subtotal 4', end=': ')
print(order4.subtotal)
```

## Usando diretamente funes

1. Nesse caso, as funes so passadas diretamente para a classe Order, sendo executada quando a property subtotal  chamada

```python
from typing import List, Callable


class Representation:

    def __str__(self):
        params = [f'{k}={v}' for k, v in self.__dict__.items()]
        return f'{self.__class__.__name__}({f", ".join(params)})'

    def __repr__(self):
        return self.__str__()
```

```python
class Product(Representation):
    def __init__(self, name: str, price: float) -> None:
        self.name = name
        self._price = price

    @property
    def price(self) -> float:
        return self._price
```

```python
class Order(Representation):

    def __init__(
            self, products: List[Product],
            discount: Callable[[float], float] = lambda x: x) -> None:
        self._total_value = sum([product.price for product in products])
        self._discount = discount

    @property
    def total(self):
        return self._total_value

    @property
    def subtotal(self):
        return self._discount(self._total_value)
```

```python
products = [
        Product('Nonebook Acer Linux', 500.0),
        Product('Keyboard Razer', 250.0),
        Product('Headphone Razer', 250.0),
    ]

order1 = Order(products, lambda x: x * 0.5)
order2 = Order(products, lambda x: x * 0.8)
order3 = Order(products)
order4 = Order(products, lambda x: x * 0.95)

print('Total 1', end=': ')
print(order1.total)
print('Total 2', end=': ')
print(order2.total)
print('Total 3', end=': ')
print(order3.total)
print('Total 4', end=': ')
print(order4.total)

print('Subtotal 1', end=': ')
print(order1.subtotal)
print('Subtotal 2', end=': ')
print(order2.subtotal)
print('Subtotal 3', end=': ')
print(order3.subtotal)
print('Subtotal 4', end=': ')
print(order4.subtotal)
```
